```python
import json
import re
import pandas as pd
import unidecode
import itertools
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
with open('data/testdata_array.json', 'r') as myfile:
    data1=myfile.read()
    
with open('data/mediavida_foro_cine.json', 'r') as myfile:
    data2=myfile.read()
    
with open('data/output.json', 'r') as myfile:
    data3=myfile.read()
```

```python
data1=re.sub(r'}\n{','},{',data1)
data1='['+data1+']'
data1 = json.loads(data1)

data2=data2[:-3]
data2 = data2+']'
data2 = json.loads(data2)

data3 = json.loads(data3)
```

```python
data=data1+data2+data3
```

```python
df = pd.DataFrame(data)
```

```python
def clean_str(s):
    
    s = unidecode.unidecode(s)
    s = re.sub(r'\t','',s)
    s = re.sub(r'\r','',s)
    s = s.lower()
    s = re.sub(r'[^a-z0-9]',' ',s)
    s = re.sub(r' +',' ',s)

    return s.strip()
```

```python
def process_summary(s):
    l = s.split('\n')
    l = list(map(clean_str,l))
    l = [i for i in l if i]
    
    return l
```

```python
df['summary'] = df['summary'].apply(lambda s: process_summary(s))
```

```python
df['spoiler_text'] = df['spoiler_text'].apply(lambda l: list(map(clean_str,l)))
```

```python
df['n_spoilers'] = df['spoiler_text'].apply(lambda l: len(l))
df['n_summary'] = df['summary'].apply(lambda l: len(l))
```

```python
df
```

```python
summaries = df.summary.tolist()
spoilers = df.spoiler_text.tolist()
```

```python
summaries = list(itertools.chain.from_iterable(summaries))
spoilers = list(itertools.chain.from_iterable(spoilers))
```

```python
summaries=[o for o in summaries if o]
spoilers = [o for o in spoilers if o]
```

```python
summaries = pd.DataFrame(summaries, columns=['text'])
spoilers = pd.DataFrame(spoilers, columns=['text'])
```

```python
summaries['spoiler'] = 0
spoilers['spoiler'] = 1
```

```python
spoilers['words'] = spoilers['text'].apply(lambda s: s.split())
```

```python
words = spoilers.words.tolist()
words = list(itertools.chain.from_iterable(words))
```

```python
spoilers['len'] = spoilers['text'].apply(lambda s: len(s.split()))
summaries['len'] = summaries['text'].apply(lambda s: len(s.split()))
```

```python
DF = pd.concat([spoilers, summaries])
DF = DF.drop_duplicates('text')
DF = DF.sample(frac=1)
DF = DF.reset_index(drop=True)
```

```python
DF = DF[['len','spoiler','text']]
```

```python
train, test = train_test_split(DF, test_size=0.2, random_state=42)
```

```python
train, val = train_test_split(train, test_size = 0.2, random_state=42)
```
